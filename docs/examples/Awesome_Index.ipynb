```python
# Example 1: Calculate NDVI (same as before)
segmentation_layer.attach_function(
    attach_ndvi,
    name="ndvi_stats",
    nir_column="band_4_mean",
    red_column="band_3_mean",
    output_column="NDVI",
)

# Example 2: Calculate multiple predefined indices (enhanced version)
segmentation_layer.attach_function(
    attach_spectral_indices,
    name="spectral_indices",
    bands={
        "blue": "band_1_mean",
        "green": "band_2_mean",
        "red": "band_3_mean",
        "nir": "band_4_mean",
    }
)

# Example 3: Calculate a single custom index with formula
segmentation_layer.attach_function(
    attach_spectral_index,
    name="ndbi_calculation",
    index_name="NDBI",
    bands={
        "SWIR1": "band_5_mean",  # If you have SWIR bands
        "NIR": "band_4_mean"
    }
)

# Example 4: Calculate custom formula
segmentation_layer.attach_function(
    attach_spectral_index,
    name="custom_vegetation_index",
    index_name="Custom_VI",
    formula="(NIR - RED) / (NIR + RED + 0.5)",
    bands={
        "NIR": "band_4_mean",
        "RED": "band_3_mean"
    }
)

# Example 5: Calculate multiple indices at once
segmentation_layer.attach_function(
    attach_multiple_spectral_indices,
    name="all_vegetation_indices",
    indices=["NDVI", "GNDVI", "SAVI", "EVI"],
    bands={
        "BLUE": "band_1_mean",
        "GREEN": "band_2_mean", 
        "RED": "band_3_mean",
        "NIR": "band_4_mean"
    }
)

# Example 6: Mixed predefined and custom indices
segmentation_layer.attach_function(
    attach_multiple_spectral_indices,
    name="mixed_indices",
    indices={
        "NDVI": None,  # Use predefined formula
        "Custom_Ratio": "(NIR + GREEN) / (RED + BLUE)",  # Custom formula
        "SAVI": None   # Use predefined formula
    },
    bands={
        "BLUE": "band_1_mean",
        "GREEN": "band_2_mean",
        "RED": "band_3_mean", 
        "NIR": "band_4_mean"
    }
)
```
